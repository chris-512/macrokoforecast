```python
import copy
df3 = pd.read_excel('data/korea_interest_rates.xlsx')
df3['Month'] = df3['Date'].str.split('월').str[0]
df3 = df3.drop(columns=['Date'])


df3
```

```python
last_year = int(df3.iloc[-1]['Year'])
last_month = int(df3.iloc[-1]['Month'])

year = 2022
month = 7

while True:
    
    found = -1
    for i, row in df3.iterrows():
        if year == int(row['Year']) and month < int(row['Month']):
            found = i+1
            break

    if found != -1:
        # df3 = df3.sort_index().reset_index(drop=True)
        data = copy.deepcopy(df3.loc[found-1])
        data['Month'] = "%02d" % month
        data['Year'] = str(year)
        print(data)
        df3 = pd.concat([df3[0:found], data, df3[found:]], ignore_index=True)
    
    if month == 1:
        month = 12
        year -= 1
    else:
        month -= 1
        
    if last_year == year and last_month == month:
        break
        
```
